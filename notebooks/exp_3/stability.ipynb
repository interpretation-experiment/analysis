## 2 Stability of sentences

First generate our data

```python
data = []
order = []
for tree in Tree.objects.experiment.all():
    root = tree.root
    root_length = len(root.content_words)
    order.append({'tree id': tree.id,
                  'root length': root_length})
    for sentence in tree.sentences.kept.all():
        if sentence.parent is None:
            raw_rate = 0.0
            raw_cumrate = 0.0
            oc_rate = 0.0
            oc_cumrate = 0.0
            ncoc_rate = 0.0
            ncoc_cumrate = 0.0
        else:
            raw_rate = sentence.raw_distance(sentence.parent)
            raw_cumrate = sentence.cum_root_distance('raw')
            oc_rate = sentence.oc_distance(sentence.parent)
            oc_cumrate = sentence.cum_root_distance('oc')
            ncoc_rate = sentence.ncoc_distance(sentence.parent)
            ncoc_cumrate = sentence.cum_root_distance('ncoc')
        length = len(sentence.content_words)
        data.append({'tree id': tree.id,
                     'sentence id': sentence.id,
                     'length': length,
                     'root length': root_length,
                     'length / len(root)': length / root_length,
                     'diff from root': root.ncoc_distance(sentence, normalized=False) / root_length,
                     'depth': sentence.depth,
                     'raw rate': raw_rate,
                     'raw cumrate': raw_cumrate,
                     'oc rate': oc_rate,
                     'oc cumrate': oc_cumrate,
                     'ncoc rate': ncoc_rate,
                     'ncoc cumrate': ncoc_cumrate})
data = DataFrame(data)
root_order = DataFrame(order).sort_values(['root length', 'tree id'])
```

```python
if SAVE_DATA:
    data.to_pickle(DATA.format('stability'))
```
